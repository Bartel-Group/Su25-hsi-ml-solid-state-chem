## Functions (`def`)
- if you want to execute any operation more than once, it may be a good idea to make a function
- in general, this is a good way to make your code readable and user-friendly
- functions take args. these args can have default values (arg=something) or not. if they don't have a default value, then the function will not run unless you provide the required args


````python
def f(x, verbose=False):
    if verbose:
        print('squaring %.2f' % float(x))
    return x**2
````

You should always write `docstrings` for functions so you (and others) know how to use them

````python
def f(x, verbose=False):
    """
    This function returns the square of a value or array

    Args:
        x (int, float, np.ndarray) : a number or array of numbers
        verbose (bool) : if True, print some helpful output

    Returns:
        the square of x
            if x is an int or float, this returns an int or float (that number squared)
            if x is an array, this returns an array where all the values in that array are squared
    """
    if verbose:
        print('squaring %.2f' % float(x))
    return x**2
````

Thermally-actived diffusion
Ion transport in crystals is a fundamental process that underpins various technological applications, from batteries to semiconductor devices. Understanding the kinetics of ion movement within and between materials is crucial for optimising device performance.

Like many chemical processes, solid-state diffusion transport is thermally activated. We can describe ion motion in a crystal using a familiar Arrhenius relationship.

The diffusion coefficient of a species is given by 
$$
D_{ion} = D_0 \cdot e^{-(\frac{\Delta E_a}{k_BT})}
$$
- $D_{ion}$ is the diffusion coefficient for a particular ion
- $D_0$ is the temperature-independent prefactor (containing an attempt frequency),
- $\Delta E_a$ is the activation energy for diffusion,
- $k_B$ is the Boltzmann constant, and
- $T$ is the temperature.
